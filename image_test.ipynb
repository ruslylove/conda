## Natural Granite Texture with Strain

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = (15,5)

img = cv2.imread('Natural-stone-Texture.jpg',2)

kernel = np.ones((50,50),np.float32)/(50*50)
dst = cv2.filter2D(img,-1,kernel)

plt.subplot(121),plt.imshow(img,cmap="gray"),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.axhline(y=250,color='red',linestyle='--')
plt.subplot(122),plt.imshow(dst,cmap="gray"),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.axhline(y=250,color='red',linestyle='--')

plt.show()
```
